```python
# загружаем пакет csv
import csv

# создаем переменные
stage = 0
flats_list = list()
flats_custom_list = list()

# открываем файл с квартирами
with open('output.csv', newline='', encoding="utf-8") as csvfile:
    flats_csv = csv.reader(csvfile, delimiter=';')
    flats_list = list(flats_csv)

# шапка
header = flats_list.pop(0)

# создаем цикл для обработки данных
while True:

# стадия 0, запрашиваем начальную информацию от пользователя
    if stage == 0:
        user_input = input("Введите ваш бюджет. Чтобы закончить, введите \"выход\"")
        if user_input.lower() == "выход":
            stage = 10
            continue
        if not user_input.isdigit():
            print("Бюджет необходимо вводить цифрами.")
            continue
        budget = int(user_input)
        stage = 1

# стадия 1, если бюдет введен правильно, ищем квартиры согласно бюджета
    if stage == 1:
        flats_custom_list.clear()
        for flat in flats_list:
            if int(flat[11]) <= budget:
                flats_custom_list.append(flat)
        if len(flats_custom_list) == 0:
            print("Для вашей суммы не найдены квартиры, попробуйте увеличить бюджет.")
            stage = 0
            continue
        user_answer = input("Количество подходящих квартир - {}.\n1. Для просмотра введите \"смотреть\".\n2. Для изменения бюджета введите новую сумму.\n3. Для выбора по метро введите список метро через запятую.".format(len(flats_custom_list)))
        if user_answer.lower() == "смотреть":
            print(flats_custom_list)
            stage = 0
            continue
        if user_answer.isdigit():
            budget = int(user_answer)
            continue
        stage = 2
        
# стадия 2, если найдены квартиры, просим ввести ближайшие метро
    if stage == 2:
        user_answer = user_answer.lower()
        subway_list = user_answer.split(",")
        subway_list = [x.strip() for x in subway_list]
        subway_dict = dict()
        subway_dict.clear()
        count_dict = 0
        for flat in flats_custom_list:
            subway_str = flat[3].replace("м.", "")
            subway_str = subway_str.lower()
            if subway_str in subway_list:
                if subway_str not in subway_dict.keys():
                    subway_dict[subway_str] = list()
                subway_dict[subway_str].append(flat)
                count_dict += 1
        if count_dict < 5:
            user_answer2 = input("По выбранному метро найдено мало квартир - {}. Попробуйте добавить еще метро через запятую.".format(count_dict))
            user_answer = user_answer2
            continue
        user_answer2 = input("По выбранному метро найдено квартир - {}.\n1. Для просмотра введите \"смотреть\".\n2. Для изменения бюджета введите новую сумму.\n3. Для выбора по метро введите новый список метро через запятую.".format(count_dict))
        if user_answer2.lower() == "смотреть":
            print(subway_dict)
            stage = 0
            continue
        if user_answer2.isdigit():
            budget = int(user_answer2)
            stage = 1
            continue
        user_answer = user_answer2
        
# стадия 10, если введено слово "выход"
    if stage == 10:
        print("Спасибо за использование нашего продукта!")
        break
```
